## [기본 사용법 및 종류](https://seaborn.pydata.org/api.html#)

| **시각화 목적** | **시각화 함수 종류** | **설명** |
|:---:|:---:|:---:|
| **Categorical plots** | countplot, barplot, boxplot, catplot, violinplot, stripplot, swarmplot, pointplot | 카테고리형 변수와 실수형 변수 간의 관계를 나타내기 위한 시각화 |
| **Matrix plots** | heatmap, clustermap | 카테고리형 변수들 간의 비율을 2차원 행렬로 만들고 그 비율에 색을 입혀 시각화 |
| **Distribution plots** | hisplot, kdeplot, rugplot, ecdfplot, distplot | 실수형 변수 하나(x) 또는 두개(x,y)의 데이터 값 분포를 파악하기 위한 시각화 |
| **Relational plots** | scatterplot, lineplot, relplot | 실수형 두가지 변수(x, y)의 관계를 나타내기 위한 시각화 |

| **Multi plots** | pairplot, jointplot | 여러 변수들간 관계를 한눈에 시각화 |

## [**Count Plot:** 카테고리 변 실수 값 빈도 집계](http://seaborn.pydata.org/generated/seaborn.countplot.html)

- `histplot`과 마찬가지로 동일한 데이터의 갯수를 `count`
- `histplot`은 실수형 데이터, `countplot`은 카테고리형 데이터를 입력받음

> - `countplot`: 카테고리 값들에 따른 데이터 갯수 표현
>
>```python
sns.countplot(x, data)    # X축 DataFrame 변수명, DataFrame


````python
# 기본 사용
plt.figure(figsize=(10,5))
sns.countplot(data=titanic, x='class')
plt.title('Titanic Data')
plt.show()
````

````python
# 기본 사용
# hue 키워드 인수 사용하여 세부 집단 추가 가능
plt.figure(figsize=(10,5))
sns.countplot(data=titanic, x='class', hue='sex')
plt.title('Titanic Data')
plt.show()
````

## [**Bar Chart:** 카테고리 별 실수 값 차이 확인](http://seaborn.pydata.org/generated/seaborn.barplot.html)

- 2개의 변수를 비교하는 `이변량(Bivariate)` 분석을 위한 시각화
- `x축`에는 카테고리 변수, `y축`에는 실수형 변수 (반대도 가능)
- 카테고리 항목에 따른 실수 값, 평균, 편차 등을 비교하는데 사용
- 막대 그래프에 오차를 기본적으로 포함하여 오차의 범위를 막대의 형식으로 표시
- **Error Bar:** 입력한 데이터 기반 유사한 상황의 95%가 오차 막대 이내의 결과를 얻을 것

>```python
sns.barplot(x, y, data)    # X축 DataFrame 변수명, Y축 DataFrame 변수명, DataFrame
```

````python
# 기본 사용
plt.figure(figsize=(10,5))
sns.barplot(data=tips, x='sex', y='tip')
plt.title('Tip Data')
plt.show()
````

````python
# 기본 사용
# estimator 키워드 인수 사용하여 표기 값 변경 가능
plt.figure(figsize=(10,5))
sns.barplot(data=tips, x='sex', y='tip', estimator=sum)
plt.title('Tip Data')
plt.show()
````

````python
# 기본 사용
# estimator 키워드 인수 사용하여 표기 값 변경 가능
plt.figure(figsize=(10,5))
sns.barplot(data=tips, x='sex', y='tip', estimator=len)
plt.title('Tip Data')
plt.show()
````

````python
# 기본 사용
# hue 키워드 인수 사용하여 세부 집단 추가 가능
plt.figure(figsize=(10,5))
sns.barplot(data=tips, x='sex', y='tip', estimator=np.mean, hue='day')
plt.title('Tip Data')
plt.show()
````

````python
# 기본 사용
# ci 키워드 인수 사용하여 에어 막대 삭제 가능
plt.figure(figsize=(10,5))
sns.barplot(data=tips, x='sex', y='tip', estimator=np.mean, hue='day', ci=None)
plt.title('Tip Data')
plt.show()
````

````python
# 기본 사용
# 축을 바꾸어 수평 막대 그래프 가능
plt.figure(figsize=(10,5))
sns.barplot(data=tips, y='sex', x='tip', estimator=np.mean, hue='day', ci=None)
plt.title('Tip Data')
plt.show()
````

## [**Box Plot:** 카테고리 별 실수 분포 확인](http://seaborn.pydata.org/generated/seaborn.boxplot.html)

- 여러 변수들의 통계값(최대, 최소, 평균, 중앙값 등)과 이상치(Outlier) 발견하는데 용이
- 데이터의 전체적인 분포를 확인하기 좋고 카테고리에 따른 실수 데이터 분포 확인
- 변수가 1개일 때는 실수형 값만 입력, 변수가 2개일 때는 카테고리와 실수를 함께 입력

---

- **Maximum:** `Q3에서 1.5 * Interquartile Range(IQR, Q3-Q1)` 위 범위까지 중 최상위
- **Quantile 25%(Q3):** 등수 상 상위 25%
- **Quantile 50%(Q2, Median):** 등수 상 가운데
- **Quantile 25%(Q1):** 등수 상 하위 25%
- **Minimum:** `Q1에서 1.5 * Interquartile Range(IQR, Q3-Q1)` 아래 범위까지 중 최하위

> <center><img src='Image/Basic/Visualization_Plot_Box2.png' width='400'></center>
>
>```python
sns.boxplot(x, y, data)    # X축 DataFrame 변수명, Y축 DataFrame 변수명, DataFrame
>```

````python
# 기본 사용
# 변수가 1개일 때
plt.figure(figsize=(10,5))
sns.boxplot(data=tips, x='tip')
plt.title('Tip Data')
plt.show()
````

````python
# 기본 사용
# 변수가 2개일 때
plt.figure(figsize=(10,5))
sns.boxplot(data=tips, x='sex', y='tip')
plt.title('Tip Data')
plt.show()
````

````python
# 기본 사용
# hue 키워드 인수 사용하여 세부 집단 추가 가능
plt.figure(figsize=(10,5))
sns.boxplot(data=tips, x='sex', y='tip', hue='day')
plt.title('Tip Data')
plt.show()
````

````python
# 기본 사용
# orient 키워드 인수 사용하여 방향 전환 가능
# x와 y의 변수도 변환해야 함
plt.figure(figsize=(10,5))
sns.boxplot(data=tips, y='sex', x='tip', hue='day', orient='h')
plt.title('Tip Data')
plt.show()
````

## [**Violin Plot:** 카테고리 별 실수 분포 확인](http://seaborn.pydata.org/generated/seaborn.violinplot.html)

- 세로 방향으로 분포를 함께 표시하며 좌우 분포가 대칭이 되어 바이올린처럼 보임

>```python

````python
# 기본 사용
# orient 키워드 인수 사용하여 방향 전환 가능
# x와 y의 변수도 변환해야 함
plt.figure(figsize=(10,5))
sns.violinplot(data=tips, y='sex', x='tip', hue='day', orient='h')
plt.title('Tip Data')
plt.show()
````

# **실수형 모든 변수 분포 및 관계 확인(Multi Plots)**

> **"여러 변수들간 관계를 한눈에 시각화"**
> - `jointplot`
> - `pairplot`


## [**Joint Plot:** 실수 변수들의 분포 및 관계 확인](http://seaborn.pydata.org/generated/seaborn.jointplot.html)

- 2차원의 데이터 표현 및 관계를 살펴보는데 많이 사용    
- `jointplot`은 Scatter Plot 뿐 아니라 가장자리에 변수의 히스토그램 분포도 표현

>```python
sns.jointplot(x, y, data)    # X축 DataFrame 변수명, Y축 DataFrame 변수명, DataFrame


````python
# 기본 사용
plt.figure(figsize=(10,10))
sns.jointplot(data=iris, kind='scatter')
plt.title('Iris Data')
plt.show()
````

````python
# 기본 사용
# kind 키워드 인수 사용하여 다양하게 표현 가능
# kind : { "scatter" | "kde" | "hist" | "hex" | "reg" | "resid" }
plt.figure(figsize=(10,10))
sns.jointplot(data=iris, kind='kde')
plt.title('Iris Data')
plt.show()
````

````python
# 기본 사용
# 특정 변수 대상으로 표현 가능
plt.figure(figsize=(10,10))
sns.jointplot(data=iris, x='sepal_length', y='sepal_width', kind='kde')
plt.title('Iris Data')
plt.show()
````

## [**Pair Plot:** 관계성 확인](http://seaborn.pydata.org/generated/seaborn.pairplot.html)

- 모든 데이터 변수 조합에 대해서 Scatter Plot
- 행과 열이 같은 변수인 경우에는 해당 변수의 히스토그램 출력

>```python
sns.pairplot(data)    # DataFrame


````python
# 기본 사용
plt.figure(figsize=(10,10))
sns.pairplot(data=iris, kind='scatter')
plt.title('Iris Data')
plt.show()
````

````python
# 기본 사용
# hue 키워드 인수 사용하여 세부 집단 추가 가능
plt.figure(figsize=(10,10))
sns.pairplot(data=iris, kind='scatter', hue='species')
plt.title('Iris Data')
plt.show()
````
